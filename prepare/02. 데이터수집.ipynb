```python
# ticker_info ÎÇ†Ïßú
from datetime import datetime
today = datetime.today().strftime('%Y-%m-%d')

# Í∏∞Î≥∏ Ï†ïÎ≥¥
ticker_info = ticker.info    

# === [1] Í∏∞Î≥∏ ÌöåÏÇ¨ Ï†ïÎ≥¥ ===
name = ticker_info.get('longName', 'N/A')
exchange = ticker_info.get('fullExchangeName', 'N/A')
industry = ticker_info.get('industry', 'N/A')
sector = ticker_info.get('sector', 'N/A')
employees = ticker_info.get('fullTimeEmployees', 'N/A')

# === [2] Ï£ºÏãù/ÏãúÍ∞ÄÏ¥ùÏï° ===
current_price = ticker_info.get('currentPrice', 'N/A')
market_cap = ticker_info.get('marketCap', 'N/A')

# === [3] Î∞∏Î•òÏóêÏù¥ÏÖò ===
trailingPE = ticker_info.get('trailingPE', 'N/A')
forwardPE = ticker_info.get('forwardPE', 'N/A')
priceToBook = ticker_info.get('priceToBook', 'N/A')
priceToSales = ticker_info.get('priceToSalesTrailing12Months', 'N/A')
pegRatio = ticker_info.get('trailingPegRatio', 'N/A')

# === [4] ÏàòÏùµÏÑ± ===
grossMargins = ticker_info.get('grossMargins', 'N/A')
operatingMargins = ticker_info.get('operatingMargins', 'N/A')
ebitdaMargins = ticker_info.get('ebitdaMargins', 'N/A')
profitMargins = ticker_info.get('profitMargins', 'N/A')
returnOnAssets = ticker_info.get('returnOnAssets', 'N/A')
returnOnEquity = ticker_info.get('returnOnEquity', 'N/A')

# === [5] Ïû¨Î¨¥ ÏïàÏ†ïÏÑ± ===
debtToEquity = ticker_info.get('debtToEquity', 'N/A')
currentRatio = ticker_info.get('currentRatio', 'N/A')
quickRatio = ticker_info.get('quickRatio', 'N/A')
totalCash = ticker_info.get('totalCash', 'N/A')
totalDebt = ticker_info.get('totalDebt', 'N/A')
freeCashflow = ticker_info.get('freeCashflow', 'N/A')

# === [6] ÏÑ±Ïû•ÏÑ± ===
earningsGrowth = ticker_info.get('earningsGrowth', 'N/A')
revenueGrowth = ticker_info.get('revenueGrowth', 'N/A')

# === [7] Î∞∞Îãπ ===
dividendRate = ticker_info.get('dividendRate', 'N/A')
dividendYield = ticker_info.get('dividendYield', 'N/A')
payoutRatio = ticker_info.get('payoutRatio', 'N/A')
```

- Î≥¥Í≥†ÏÑú ÏÉùÏÑ± ÌÉ¨ÌîåÎ¶ø

```python

# ÎßàÌÅ¨Îã§Ïö¥ ÌòïÏãùÏùò Î≥¥Í≥†ÏÑú ÏÉùÏÑ±
report = f"""
# {name} ({ticker}) Í∏∞ÏóÖ Ï†ïÎ≥¥ Î≥¥Í≥†ÏÑú

**ÏûëÏÑ±Ïùº**: {today}

---
## 1. üè¢ ÌöåÏÇ¨ Ï†ïÎ≥¥
- **Ï¢ÖÎ™©Î™Ö**: {name}
- **Í±∞ÎûòÏÜå(ÏãúÏû•)**: {exchange}
- **ÏÇ∞ÏóÖ**: {industry}
- **ÏÑπÌÑ∞**: {sector}
- **Ï¥ù ÏßÅÏõê Ïàò**: {employees}

---

## 2. üí∞ Ï£ºÏãù Î∞è Í∏∞ÏóÖ Í∞ÄÏπò
- **ÏµúÍ∑º Ï£ºÍ∞Ä**: {current_price} USD
- **ÏãúÍ∞ÄÏ¥ùÏï°**: {market_cap:,} USD

---

## 3. üìä Î∞∏Î•òÏóêÏù¥ÏÖò ÏßÄÌëú
- **PER (Trailing)**: {trailingPE}
- **PER (Forward)**: {forwardPE}
- **PBR**: {priceToBook}
- **PSR**: {priceToSales}
- **PEG Ratio**: {pegRatio}

---

## 4. üíπ ÏàòÏùµÏÑ± ÏßÄÌëú
- **Îß§Ï∂úÏ¥ùÏù¥ÏùµÎ•†(Gross Margin)**: {grossMargins}
- **ÏòÅÏóÖÏù¥ÏùµÎ•†(Operating Margin)**: {operatingMargins}
- **EBITDA ÎßàÏßÑ**: {ebitdaMargins}
- **ÏàúÏù¥ÏùµÎ•†(Profit Margin)**: {profitMargins}
- **ROA (ÏûêÏÇ∞ÏàòÏùµÎ•†)**: {returnOnAssets}
- **ROE (ÏûêÍ∏∞ÏûêÎ≥∏Ïù¥ÏùµÎ•†)**: {returnOnEquity}

---

## 5. üõ°Ô∏è Ïû¨Î¨¥ ÏïàÏ†ïÏÑ± ÏßÄÌëú
- **Î∂ÄÏ±ÑÎπÑÏú®(Debt to Equity)**: {debtToEquity}
- **Ïú†ÎèôÎπÑÏú®(Current Ratio)**: {currentRatio}
- **ÎãπÏ¢åÎπÑÏú®(Quick Ratio)**: {quickRatio}
- **ÌòÑÍ∏à Î≥¥Ïú†Îüâ**: {totalCash:,} USD
- **Ï¥ù Î∂ÄÏ±Ñ**: {totalDebt:,} USD
- **ÏûâÏó¨ÌòÑÍ∏àÌùêÎ¶Ñ (FCF)**: {freeCashflow:,} USD

---

## 6. üìà ÏÑ±Ïû•ÏÑ± ÏßÄÌëú
- **ÏàúÏù¥Ïùµ ÏÑ±Ïû•Î•†**: {earningsGrowth}
- **Îß§Ï∂ú ÏÑ±Ïû•Î•†**: {revenueGrowth}

---

## 7. üíµ Î∞∞Îãπ Í¥ÄÎ†® ÏßÄÌëú
- **Ïó∞Í∞Ñ Î∞∞ÎãπÍ∏à**: {dividendRate} USD
- **Î∞∞Îãπ ÏàòÏùµÎ•†**: {dividendYield}
- **Î∞∞Îãπ ÏÑ±Ìñ•**: {payoutRatio}

"""
```

```python
# Î≥¥Í≥†ÏÑú Ï∂úÎ†•
print(report)
```

```python
# ÌååÏùºÎ™Ö ÏÉùÏÑ± (Ïòà: AAPL_2025-06-11_report.md)
filename = f"{symbol}_{today}_report.md"

# ÌååÏùºÎ°ú Ï†ÄÏû•
with open(filename, "w", encoding="utf-8") as f:
    f.write(report)

print(f"‚úÖ Î≥¥Í≥†ÏÑúÍ∞Ä ÏÑ±Í≥µÏ†ÅÏúºÎ°ú Ï†ÄÏû•ÎêòÏóàÏäµÎãàÎã§: {filename}")
```
