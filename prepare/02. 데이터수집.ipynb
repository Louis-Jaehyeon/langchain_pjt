# 데이터 수집

```python
import yfinance as yf

# 1. 애플(AAPL)의 주식 티커 데이터 가져오기 (JSON 형태)
ticker = 'AAPL'
stock = yf.Ticker(ticker)

stock_info = stock.info
stock_info
```

```python
# 2. 주가 데이터 가져오기 (최근 1년 간의 일일 주가 데이터, DataFrame 형태)
stock.history(period='1y')
```

```python
# stock_info['shortName']
stock_info.get('shortName')
```

```python
# 현재 날짜
from datetime import datetime
today = datetime.today().strftime('%Y-%m-%d')

# 기본 정보
stock_info = stock.info    

# === [1] 기본 회사 정보 ===
name = stock_info.get('longName', 'N/A')
exchange = stock_info.get('fullExchangeName', 'N/A')
industry = stock_info.get('industry', 'N/A')
sector = stock_info.get('sector', 'N/A')
employees = stock_info.get('fullTimeEmployees', 'N/A')

# === [2] 주식/시가총액 ===
current_price = stock_info.get('currentPrice', 'N/A')
market_cap = stock_info.get('marketCap', 'N/A')

# === [3] 밸류에이션 ===
trailingPE = stock_info.get('trailingPE', 'N/A')
forwardPE = stock_info.get('forwardPE', 'N/A')
priceToBook = stock_info.get('priceToBook', 'N/A')
priceToSales = stock_info.get('priceToSalesTrailing12Months', 'N/A')
pegRatio = stock_info.get('trailingPegRatio', 'N/A')

# === [4] 수익성 ===
grossMargins = stock_info.get('grossMargins', 'N/A')
operatingMargins = stock_info.get('operatingMargins', 'N/A')
ebitdaMargins = stock_info.get('ebitdaMargins', 'N/A')
profitMargins = stock_info.get('profitMargins', 'N/A')
returnOnAssets = stock_info.get('returnOnAssets', 'N/A')
returnOnEquity = stock_info.get('returnOnEquity', 'N/A')

# === [5] 재무 안정성 ===
debtToEquity = stock_info.get('debtToEquity', 'N/A')
currentRatio = stock_info.get('currentRatio', 'N/A')
quickRatio = stock_info.get('quickRatio', 'N/A')
totalCash = stock_info.get('totalCash', 'N/A')
totalDebt = stock_info.get('totalDebt', 'N/A')
freeCashflow = stock_info.get('freeCashflow', 'N/A')

# === [6] 성장성 ===
earningsGrowth = stock_info.get('earningsGrowth', 'N/A')
revenueGrowth = stock_info.get('revenueGrowth', 'N/A')

# === [7] 배당 ===
dividendRate = stock_info.get('dividendRate', 'N/A')
dividendYield = stock_info.get('dividendYield', 'N/A')
payoutRatio = stock_info.get('payoutRatio', 'N/A')
```

- 보고서 생성 탬플릿

```python

# 마크다운 형식의 보고서 생성
report = f"""
# {name} ({ticker}) 기업 정보 보고서

**작성일**: {today}

---
## 1. 🏢 회사 정보
- **종목명**: {name}
- **거래소(시장)**: {exchange}
- **산업**: {industry}
- **섹터**: {sector}
- **총 직원 수**: {employees}

---

## 2. 💰 주식 및 기업 가치
- **최근 주가**: {current_price} USD
- **시가총액**: {market_cap:,} USD

---

## 3. 📊 밸류에이션 지표
- **PER (Trailing)**: {trailingPE}
- **PER (Forward)**: {forwardPE}
- **PBR**: {priceToBook}
- **PSR**: {priceToSales}
- **PEG Ratio**: {pegRatio}

---

## 4. 💹 수익성 지표
- **매출총이익률(Gross Margin)**: {grossMargins}
- **영업이익률(Operating Margin)**: {operatingMargins}
- **EBITDA 마진**: {ebitdaMargins}
- **순이익률(Profit Margin)**: {profitMargins}
- **ROA (자산수익률)**: {returnOnAssets}
- **ROE (자기자본이익률)**: {returnOnEquity}

---

## 5. 🛡️ 재무 안정성 지표
- **부채비율(Debt to Equity)**: {debtToEquity}
- **유동비율(Current Ratio)**: {currentRatio}
- **당좌비율(Quick Ratio)**: {quickRatio}
- **현금 보유량**: {totalCash:,} USD
- **총 부채**: {totalDebt:,} USD
- **잉여현금흐름 (FCF)**: {freeCashflow:,} USD

---

## 6. 📈 성장성 지표
- **순이익 성장률**: {earningsGrowth}
- **매출 성장률**: {revenueGrowth}

---

## 7. 💵 배당 관련 지표
- **연간 배당금**: {dividendRate} USD
- **배당 수익률**: {dividendYield}
- **배당 성향**: {payoutRatio}

"""
```

```python
# 보고서 출력
print(report)
```

```python
# 파일명 생성 (예: AAPL_2025-06-11_report.md)
filename = f"{ticker}_{today}_report.md"

# 파일로 저장
with open(filename, "w", encoding="utf-8") as f:
    f.write(report)

print(f"✅ 보고서가 성공적으로 저장되었습니다: {filename}")
```

```python
# 최근 6개월 간의 주가 데이터를 가져오기 (기간 설정 가능: "1mo", "3mo", "6mo", "1y" 등)
# interval 옵션은 데이터 수집 간격을 의미 (e.g., "1d": 일간 데이터, "1wk": 주간 데이터, "1mo": 월간 데이터)
hist = stock.history(period="6mo", interval="1d")

# 주가 데이터 출력 (OHLC 및 거래량)
print(hist[['Open', 'High', 'Low', 'Close', 'Volume']])
# print(type(hist))
```
